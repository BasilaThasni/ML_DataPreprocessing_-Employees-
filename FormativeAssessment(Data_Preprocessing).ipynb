```python
df.rename(columns = {"Age":"Employee's Age","Salary":"Employee's Salary",},inplace = True)
df
```

```python
df["Company"]=df["Company"].replace("Infosys Pvt Lmt","Infosys")
df["Company"]=df["Company"].replace("Tata Consultancy Services","TCS")
df["Company"]=df["Company"].replace("Congnizant","CTS")
df

```

##DATA CLEANING:

```python
df.isnull().sum()
```

```python
import numpy as np
for column in["Company","Place"]:
  df[column].fillna("Unknown",inplace=True)
df.isnull().sum()
```

##Replace the value 0 in age as NaN.

```python
df["Employee's Age"].replace(0, np.NaN, inplace=True)
```

```python
df
```

```python
df.fillna({
    "Employee's Age": df["Employee's Age"].median(),
    "Employee's Salary": df["Employee's Salary"].mean(),
    }, inplace=True)
df.isnull().sum()
```

```python
df.drop_duplicates(inplace=True)
dataframe = df
df = df.reset_index(drop=True)
df.duplicated().sum()
```

```python
df.shape
```

#DATA ENCODING:

###Convert categorical variables into numerical representations using techniques such as one-hot encoding, label encoding, making them suitable for analysis by machine learning algorithms.

```python
from sklearn.preprocessing import OneHotEncoder
# One-Hot Encoding for multi-category columns
encoder = OneHotEncoder(sparse_output=False)
encoded_columns = encoder.fit_transform(df[['Company', 'Place']])
encoded_features = encoder.get_feature_names_out(['Company', 'Place'])
encoded_df = pd.DataFrame(encoded_columns, columns=encoded_features)
encoded_df
```

```python
final_df = pd.concat([df[["Employee's Age","Employee's Salary",'Gender']], encoded_df], axis=1)
final_df
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
new_df = df.copy()
for column in ["Company","Place"]:
  new_df[column]=label_encoder.fit_transform(new_df[column])
new_df
```

##No use of encoding 'Country' column since it is not going to be used.

#FEATURE SCALING:

##perform the scaling of the features using standardscaler and minmaxscaler.


```python
from sklearn.preprocessing import StandardScaler,MinMaxScaler
stdscalar = StandardScaler()
final_df[["Employee's Age","Employee's Salary"]]=stdscalar.fit_transform(final_df[["Employee's Age","Employee's Salary"]])
final_df
```

```python
minmaxsc = MinMaxScaler()
new_df[["Employee's Age","Employee's Salary"]] = minmaxsc.fit_transform(new_df[["Employee's Age","Employee's Salary"]])
new_df
```
